```python
%%sql
CREATE FUNCTION fn_check_adjacent_repetition(p_str VARCHAR(20))
RETURNS TINYINT
DETERMINISTIC
BEGIN
    DECLARE i INT DEFAULT 1;
    DECLARE str_len INT;

    SET str_len = CHAR_LENGTH(p_str);

    WHILE i < str_len DO
        IF SUBSTRING(p_str, i, 1) = SUBSTRING(p_str, i+1, 1) THEN
            RETURN 1; -- TRUE
        END IF;
        SET i = i + 1;
    END WHILE;

    RETURN 0; -- FALSE
END;

```

```python
%%sql
DROP FUNCTION IF EXISTS fn_check_adjacent_repetition;
```

```python
%%sql 
SELECT Pan_numbers,
       fn_check_adjacent_repetition(Pan_numbers) AS has_adjacent_repetition
FROM Pan
limit 10;
```

```python
%%sql
CREATE FUNCTION fn_check_sequence(p_str VARCHAR(20))
RETURNS TINYINT
DETERMINISTIC
BEGIN
    DECLARE i INT DEFAULT 1;
    DECLARE str_len INT;

    SET str_len = CHAR_LENGTH(p_str);

    -- Loop through each character
    WHILE i < str_len DO
        IF ASCII(SUBSTRING(p_str, i+1, 1)) - ASCII(SUBSTRING(p_str, i, 1)) <> 1 THEN
            RETURN 0; -- FALSE (not sequential)
        END IF;
        SET i = i + 1;
    END WHILE;

    RETURN 1; -- TRUE (all characters sequential)
END;

```

```python
%%sql 
SELECT Pan_numbers,
       fn_check_sequence(Pan_numbers) AS has_adjacent_repetition
FROM Pan
limit 10;
```

```python
%%sql
CREATE OR REPLACE VIEW vw_valid_invalid_pans AS
SELECT cln.pan_numbers,
       CASE 
           WHEN vld.pan_numbers IS NULL THEN 'Invalid PAN'
           ELSE 'Valid PAN'
       END AS status
FROM (
    -- Clean PAN numbers: trim, uppercase, remove internal spaces
    SELECT DISTINCT UPPER(TRIM(REPLACE(pan_numbers, ' ', ''))) AS pan_numbers
    FROM pan
    WHERE pan_numbers IS NOT NULL
      AND TRIM(pan_numbers) <> ''
) AS cln
LEFT JOIN (
    -- Apply validation rules
    SELECT pan_numbers
    FROM (
        SELECT DISTINCT UPPER(TRIM(REPLACE(pan_numbers, ' ', ''))) AS pan_numbers
        FROM pan
        WHERE pan_numbers IS NOT NULL
          AND TRIM(pan_numbers) <> ''
    ) AS base
    WHERE fn_check_adjacent_repetition(SUBSTRING(pan_numbers,1,5)) = 0   -- no adjacent letters
      AND fn_check_sequence(SUBSTRING(pan_numbers,1,5)) = 0               -- no sequential letters
      AND fn_check_adjacent_repetition(SUBSTRING(pan_numbers,6,4)) = 0    -- no adjacent digits
      AND fn_check_sequence(SUBSTRING(pan_numbers,6,4)) = 0               -- no sequential digits
      AND pan_numbers REGEXP '^[A-Z][A-Z][A-Z][A-Z][A-Z][0-9][0-9][0-9][0-9][A-Z]$'
) AS vld
  ON vld.pan_numbers = cln.pan_numbers;
```

```python
%%sql
SELECT * FROM vw_valid_invalid_pans
limit 15;
```

```python
%%sql
    WITH cte AS (
        SELECT 
            (SELECT COUNT(*) FROM pan) AS total_processed_records,
            SUM(CASE WHEN vw.status = 'Valid PAN' THEN 1 ELSE 0 END) AS total_valid_pans,
            SUM(CASE WHEN vw.status = 'Invalid PAN' THEN 1 ELSE 0 END) AS total_invalid_pans
        FROM vw_valid_invalid_pans vw
    )
    SELECT 
        total_processed_records, 
        total_valid_pans, 
        total_invalid_pans,
        total_processed_records - (total_valid_pans + total_invalid_pans) AS missing_incomplete_pans
    FROM cte;
```
